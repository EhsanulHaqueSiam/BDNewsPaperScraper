# 📰 Bangladesh News Articles - Starter Notebook

This notebook demonstrates how to work with the **Bangladesh News Articles Dataset** - a daily-updated collection of news from 74+ Bangladeshi newspapers.

**Dataset DOI:** [10.34740/KAGGLE/DS/9384161](https://doi.org/10.34740/KAGGLE/DS/9384161)

---

## 📂 Load the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

# Load data (choose your preferred format)
df = pd.read_parquet('/kaggle/input/bangladesh-news-articles/articles.parquet')
# Alternative: df = pd.read_csv('/kaggle/input/bangladesh-news-articles/articles.csv')

print(f"📊 Total articles: {len(df):,}")
df.head()
```

## 📋 Dataset Overview

```python
# Basic info
print("📋 Column Info:")
print(df.dtypes)
print(f"\n📅 Date Range: {df['publication_date'].min()} to {df['publication_date'].max()}")
print(f"📰 Unique Newspapers: {df['paper_name'].nunique()}")
print(f"🏷️ Unique Categories: {df['category'].nunique()}")
```

## 📰 Articles by Newspaper

```python
# Top newspapers
paper_counts = df['paper_name'].value_counts().head(15)

fig, ax = plt.subplots(figsize=(12, 6))
paper_counts.plot(kind='barh', ax=ax, color=sns.color_palette('viridis', len(paper_counts)))
ax.set_xlabel('Number of Articles')
ax.set_ylabel('Newspaper')
ax.set_title('📰 Top 15 Newspapers by Article Count')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```

## 🏷️ Articles by Category

```python
# Top categories
category_counts = df['category'].value_counts().head(10)

fig, ax = plt.subplots(figsize=(10, 6))
category_counts.plot(kind='pie', ax=ax, autopct='%1.1f%%', startangle=90)
ax.set_ylabel('')
ax.set_title('🏷️ Top 10 Categories')
plt.tight_layout()
plt.show()
```

## 📝 Text Analysis

```python
# Article length distribution
df['article_length'] = df['article_body'].fillna('').str.len()
df['headline_length'] = df['headline'].fillna('').str.len()

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Article body length
axes[0].hist(df['article_length'], bins=50, color='steelblue', edgecolor='white', alpha=0.7)
axes[0].set_xlabel('Character Count')
axes[0].set_ylabel('Frequency')
axes[0].set_title('📄 Article Body Length Distribution')
axes[0].axvline(df['article_length'].median(), color='red', linestyle='--', label=f'Median: {df["article_length"].median():,.0f}')
axes[0].legend()

# Headline length
axes[1].hist(df['headline_length'], bins=50, color='coral', edgecolor='white', alpha=0.7)
axes[1].set_xlabel('Character Count')
axes[1].set_ylabel('Frequency')
axes[1].set_title('📝 Headline Length Distribution')
axes[1].axvline(df['headline_length'].median(), color='red', linestyle='--', label=f'Median: {df["headline_length"].median():,.0f}')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 🔍 Sample Articles

```python
# Display a random sample
sample = df.sample(3)[['headline', 'category', 'paper_name', 'publication_date']]
for idx, row in sample.iterrows():
    print(f"📰 {row['paper_name']} | {row['category']}")
    print(f"   📅 {row['publication_date']}")
    print(f"   📝 {row['headline'][:100]}...")
    print()
```
